## Исследование задачи

Избавимся от ненужных столбцов:

```python
df = df.drop(['row_number', 'surname', 'customer_id'], axis = 1)
```

Сформируем признаки и целевую переменную:

```python
features = df.drop('exited', axis = 1)
target = df['exited']
```

Посмотрим на баланс классов:

```python
target.value_counts(normalize = True)
```

Видим, что в цеелой переменной есть диссбаланс классов, ответов 0 в 4 раза больше, чем ответов 1. Далее проверим модели, не учитывая дисбаланс.

Используем технику one hot encoder, чтобы перевести категориальные признаи в численные

```python
features = pd.get_dummies(features, drop_first = True)
```

Разобьем данные на три выборки:

```python
features_train, features_valid, target_train, target_valid = train_test_split(
    features, target, random_state = 99, test_size = 0.4)
features_valid, features_test, target_valid, target_test = train_test_split(
    features_valid, target_valid, test_size = 0.5, random_state = 99)
```

```python
scaler = StandardScaler()
columns = features.columns
scaler.fit(features_train)
features_train[columns] = scaler.transform(features_train[columns])
features_valid[columns] = scaler.transform(features_valid[columns])
features_test[columns] = scaler.transform(features_test[columns])
```

__Bывод:__ Разделили данные на три выборки: тренировочная, валидационная и тестовая с соотношением 6:2:2, также преминили для признаков технику One-hot-encoder и стандартизировали данные с помощью StandardScaler.

## Борьба с дисбалансом

Применим два способа борьбы с дисбалансом: первый  - увеличение выборки с наименьшим классом, второе - в модели напишем class_weight = 'balanced' и проверим результаты

```python
target_train.value_counts(normalize = True)
```

функция для увеличения объектов с целевым признаком равным 1.

```python
def upsample(features, target, repeat):
    
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=99)
    
    return features_upsampled, target_upsampled

features_upsampled, target_upsampled = upsample(features_train, target_train, 4)
```
